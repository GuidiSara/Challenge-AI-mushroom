# Separated dataset from labels

```python
features = train.columns.drop('class')
x_train = train[features]
y_train = train['class']
```

```python
x_test = test[features]
y_test = test['class']
```

# Accuracy

Decision tree

```python
tree = DecisionTreeClassifier()
tree.fit(x_train, y_train)
tree.fit(x_test, y_test)
print(f'Accuracy of training set: {tree.score(x_train, y_train)*100}%')
print(f'Accuracy of test set: {tree.score(x_train, y_train)*100}%')
```

SVC

```python
svc = SVC()
svc.fit(x_train, y_train)
y_pred = svc.predict(x_test)
acc_svc_training = round(accuracy_score(y_pred, y_test) * 100, 2)

svc.fit(x_test, y_test)
y_pred = svc.predict(x_test)
acc_svc_test = round(accuracy_score(y_pred, y_test) * 100, 2)

print(f'Accuracy of Training set: {acc_svc_training} %')
print(f'Accuracy of Test set: {acc_svc_test} %')
```
